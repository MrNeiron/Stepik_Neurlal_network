```python
import numpy as np

w = np.array([0,0,0])

print("w shape", w.shape)

examples = np.array([[1,1,0.3],[1,0.4,0.5],[1,0.7,0.8]])

def Predict(ex):
    sum = ex[0] * w[0] + ex[1] * w[1] + ex[2] * w[2]
    #sum = w.dot(ex)
    print("example: ", ex)
    print("sum: ", sum)
    if (sum > 0):
        return 1
    else:
        return 0

def Target(ex):
    if (ex[1] == 1 and ex[2] == 0.3):
        return 1
    elif (ex[1] == 0.4 and ex[2] == 0.5):
        return 1
    elif (ex[1] == 0.7 and ex[2] == 0.8):
        return 0

perfect = False
while perfect != True:
    perfect = True
    for e in examples:
        print('e',e,'\n')
        if (Predict(e) != Target(e)):
            perfect = False
            if (Predict(e) == 0):
                w = w + e
            else:
                w = w - e
            print('w',w,'\n')
print("answer: ", w)
```
