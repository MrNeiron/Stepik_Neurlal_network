```python
import numpy as np


x_shape = tuple(map(int, input().split()))
X = np.fromiter(map(float, input().split()), np.float).reshape(x_shape)
print(X)
y_shape = int(input())
Y = np.fromiter(map(float, input().split()), np.float).reshape(y_shape,1)
print(Y)

W = [0,0,0]
print(W)
```

```python

perfect = False
while perfect != True:
    k = 0
    perfect = True
    for e in X:
        answer = np.sum(e * W, 0)
        print("e: ", e)
        print("answer: " , answer)
        print("Y[", k, "][0]: ", Y[k][0])
        if (answer > 0): answer = 1
        else: answer = 0
        if answer != Y[k][0]:
            perfect = False
            if answer == 0:
                W = W + e
            else:
                W = W - e
        print("w: ", W)      
        k += 1
    print(W)
    
```
